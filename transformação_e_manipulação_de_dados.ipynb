# Preparação dos dados 

```python
import pandas as pd 

dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
# estão aninhados em dicionarios 
```

```python
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
# Alguma var esta agrupados em lista 
# vamos desagrupar essas var 
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:]) # dps do indice 3 pra frente 
dados.head()
# Agr nossos dados esta desagrupado 
```

```python
dados.reset_index(inplace = True, drop = True)
dados.head()
```

```python
dados.info()
# dados do tipo object 
# . Varios tipos de dados diferente como ter um numero no texto entre outras coisas 
# . um tipo unico, mas esse tipo n foi indentificado e o arquivo colocou como object
# . Todos os tipos são strings 
```

# Convertendo dados numéricos 

```python
import numpy as np

dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64) # tipo inteiro 
# a gente alterou o tipo, isso pode ajudar na analise estatistica 
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']

dados[col_numericas] = dados[col_numericas].astype(np.int64)
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)

dados.info()

# float - é tipo numerico, exemplo: 10.0 0.25 
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip()) # substituindo em valores em brancos '$' ',' 
dados.head()
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
dados.info()
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip()) # applymap leu var por var 
dados.head()
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
dados.info()
```

# Manipulação de texto 

```python
dados['descricao_local'] = dados['descricao_local'].str.lower() # saiu nossas frases em letras minuscula 
dados.head()
```

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ',regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ',regex=True)
dados.head()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split() # metodo capaz de dividir os elementos textuais em listas
dados.head()
```

```python
dados['comodidades']= dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
dados.head()
```

```python
dados['comodidades']= dados['comodidades'].str.split(',')
dados.head()
```
